```python
def tax(num):
  if num < 10000:
    tax = 0
  elif num < 50000:
    tax = num * 0.02
  elif num < 100000:
    tax = num * 0.05
  else:
    tax = num * 0.10
  return tax

amount = int(input("Enter an amount: "))

print(tax(amount))
```
